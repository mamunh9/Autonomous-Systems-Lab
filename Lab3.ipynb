## Part 3: A tiny dataset

In machine learning, a dataset often contains:
- **samples**: individual observations,
- **features**: measured properties,
- **labels**: desired outputs.

Example:
- sample: one student,
- features: study hours, attendance,
- label: passed exam or not.

```python
X = np.array([
    [2, 1],
    [4, 1],
    [6, 0],
    [8, 1]
])

y = np.array([0, 0, 1, 1])

print("X =")
print(X)
print("y =")
print(y)
```

In this example:
- each row in `X` is one sample,
- each column in `X` is one feature,
- `y` contains the labels.
