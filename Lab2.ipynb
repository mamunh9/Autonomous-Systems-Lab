## Part 1: Indentation instead of braces

In C/C++, blocks are often written with braces `{}`.  
In Python, blocks are defined by indentation.

Run the next cell.

````python
for i in range(1,5):
    print(i)
````

## Part 3: Conditions

Conditions in Python use `if`, `elif`, and `else`.

````python
x = 2

if x > 5:
    print("x is greater than 5")
else:
    print("x is not greater than 5")
````

## Part 4: Loops over lists

In Python, you often loop directly over values instead of using array indices.

````python
values = [10, 20, 30]

for v in values:
    print(v)
````

## Part 5: Functions

A function groups code and makes it reusable.

````python
def square(x):
    return x * x

print(square(4))
````
